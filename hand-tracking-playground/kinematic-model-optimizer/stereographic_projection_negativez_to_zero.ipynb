```python
from math import *


def make_unit(x, y, z):
  len = sqrt((x**2)+(y**2)+(z**2))
  new_x = x/len
  new_y = y/len
  new_z = z/len
  return new_x, new_y, new_z


def stereographic_2d_to_unit(X, Y):
  denom = 1/(1+(X**2)+(Y**2))
  x = 2*X*denom
  y = 2*Y*denom
  z = (-1+(X**2)+(Y**2))*denom
  return (x, y, z)


def unit_to_stereographic(x,y,z):
  X = x/(1-z)
  Y = y/(1-z)
  return X, Y

vec = (0, 0.05, -1)

vec = make_unit(*vec)

stereo = unit_to_stereographic(*vec)

vec_recovered = stereographic_2d_to_unit(*stereo)

print(vec)
print(vec_recovered)
print(stereo)
```
