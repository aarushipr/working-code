```python
%matplotlib widget
import matplotlib.pyplot as plt
import scipy.optimize
import numpy as np
import pandas
import statistics

e = pandas.pandas.read_csv("timings_double.csv")

y = []

for i in e.iloc:
    # print(i[0])
    y.append( i[0] )

print(statistics.median(y))    
print(statistics.mean(y))
    
plt.figure()
plt.hist(y, alpha=0.3, bins=50)

plt.legend()
```
