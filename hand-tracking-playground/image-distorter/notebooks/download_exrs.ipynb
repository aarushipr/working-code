```python
import requests
import random
```

```python
r = requests.get("https://api.polyhaven.com/categories/hdris", "/types")
r.json()
```

```python
r = requests.get("https://api.polyhaven.com/assets?t=hdris&c=indoor", "/types")
r.json().keys()
```

```python
indoor_tags = list(r.json().keys())
```

```python
# It's ok to run this, doesn't take very long
urls = []
for tag in indoor_tags:
  r = requests.get(f"https://api.polyhaven.com/files/{tag}").json()["hdri"]["4k"]["exr"]["url"]
  # print(r)
  urls.append(r)
random.shuffle(urls)
```

```python
# Be really careful running this
for url in urls:
  fn = url.rsplit("/")[-1]
  print(fn)
  request = requests.get(url)

  with open(f"/3/epics/artificial_data_3/hdris/{fn}", "wb") as f:
    f.write(request.content)
```
